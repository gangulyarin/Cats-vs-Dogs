```python
import os
import cv2
import numpy as np
```

```python
import tensorflow as tf
import imutils

def load(imagePaths):
    data = []
    labels = []

    for (i,imagePath) in enumerate(imagePaths):
        image = cv2.imread(imagePath)
        #label = imagePath.split(os.path.sep)[-2]
        imageFileName = imagePath.split(os.path.sep)[-1]
        label = imageFileName.split('.')[0]
        dH=0
        dW=0
        (w,h) = image.shape[:2]
        if w<h:
            image = imutils.resize(image,width=32,inter=cv2.INTER_AREA)
            dH = int((image.shape[0]-32)/2.0)
        else:
            image = imutils.resize(image,height=32,inter=cv2.INTER_AREA)
            dW = int((image.shape[1]-32)/2.0)
        (h, w) = image.shape[:2]
        image = image[dH:h - dH, dW:w - dW]
        image = cv2.resize(image,(32,32),interpolation=cv2.INTER_AREA)
        image = tf.keras.preprocessing.image.img_to_array(image)
        data.append(image)
        labels.append(label)

    return (np.array(data), np.array(labels))
```

```python
from imutils import paths

imagePaths = list(paths.list_images("train/train"))
imagePaths
```

```python
(data,labels) = load(imagePaths)
labels
```
